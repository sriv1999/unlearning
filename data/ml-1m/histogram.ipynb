```python
import pandas as pd

user_partition = pd.read_pickle('C_U_type-2_num-10.pk')
movie_partition = pd.read_pickle('C_I_type-1_num-10.pk')

with open('users18_21.txt') as file:
    user_remove = [int(line.rstrip()) for line in file]

with open('movies_1990.txt') as file:
    movie_remove = [int(line.rstrip()) for line in file]

user_partition
```

```python
user_counts = {}
for i in range(0, len(user_partition)):
    user_counts[i] = 0

for user in user_remove:
    index = 0
    while index < len(user_partition):
        current_partition = user_partition[index]
        if user in current_partition:
            user_counts[index] += 1
            break
        
        index += 1

user_counts
```

```python
movie_counts = {}
for i in range(0, len(movie_partition)):
    movie_counts[i] = 0

print(movie_partition)
print(movie_remove)

for movie in movie_remove:
    index = 0
    while index < len(movie_partition):
        current_partition = movie_partition[index]
        if movie in current_partition:
            movie_counts[index] += 1
            break
        
        index += 1

movie_counts
```

```python
import matplotlib.pyplot as plt

userList = [key for key, val in user_counts.items() for _ in range(val)]

plt.hist(userList, bins=10)

plt.xlabel('Shards') 
plt.ylabel('User Count') 
plt.title(label="Users to Be Removed Distribution")

plt.show()
```

```python
plt.bar(list(user_counts.keys()), user_counts.values())

plt.xlabel('Shards') 
plt.ylabel('User Count') 
plt.title(label="Users to Be Removed Distribution")

plt.show()
```

```python
movieList = [key for key, val in movie_counts.items() for _ in range(val)]

plt.hist(movieList, bins=10, color='orange')

plt.xlabel('Shards') 
plt.ylabel('Movie Count') 
plt.title(label="Movies to Be Removed Distribution")

plt.show()
```

```python
plt.bar(list(movie_counts.keys()), movie_counts.values(), color='orange')

plt.xlabel('Shards') 
plt.ylabel('Movie Count') 
plt.title(label="Movies to Be Removed Distribution")

plt.show()
```
